```python
def grade(v):
    if v>=90:
        return '優'
    elif v>=80:
        return '甲'
    elif v>=70:
        return '乙'
    elif v>=60:
        return '乙'
    else:
        return '丁'
    
frame['評比'] = frame['平均'].map(grade)
frame
```
