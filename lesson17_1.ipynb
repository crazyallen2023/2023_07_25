```python
import pandas as pd

data = {"字串":['台北','台中'],
        "整數":[1000,2000],
        '浮點數':[3.258,7.999]}

df = pd.DataFrame(data)
display(df)
```

```python
# 視覺美化
#pip install jinja2

df.style\
.format(precision=2,thousands=',',decimal='.')\
.relabel_index(['列1','列2'])
```

```python
import numpy as np

data = np.random.rand(10,2)  * 10

df = pd.DataFrame(data,
                  index = pd.date_range(start = "2023-09-01", periods=10),
                  columns=["Taipei","Kaoshiung"])
df
```

```python
# 視覺美化資料表
def feel_condition(v):
    if v < 1.75:
        return "溫暖"
    elif v < 2.75:
        return "潮濕"
    else:
        return "日照"
    
df.style\
.format(feel_condition)\
.set_caption("體感溫度")\
.background_gradient(cmap = "PuBuGn", vmin=1, vmax=5,axis=None)
```

```python
# 隱藏資料
df = pd.DataFrame(np.random.randn(5,5))
df.style\
.hide(subset=[0, 2, 4],axis=0)\
.hide(subset=[1,3],axis=1)
```

```python
df = pd.DataFrame(np.random.randn(10,4),
             columns=["A", "B", "C", "D"])
display(df)

def style_negative(v, props=''):
    return props if v < 0 else None

df.style\
.map(style_negative,props='color:red;')
```
