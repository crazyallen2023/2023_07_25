```python
import random

def get_score()->list:
    score = []
    for i in range(5):
        score.append(random.randint(50,100))
    return score


def get_names(nums:int) -> list:
    with open('names.txt',encoding='utf-8',newline='') as file:
        names_str = file.read()
        all_names_list = names_str.split(sep="\n")
        names_list = random.choices(all_names_list,k=nums) #取出一定數量的姓名
    return names_list

nums_int = int(input("輸入學生人數："))
name_list = get_names(nums_int)
# print(type(name_list))
student = []
for i in range(nums_int):
    score = get_score()
    # print(type(score))
    namescore_list = [name_list[i]] + score
    student.append(namescore_list)

student
```

```python
import pandas as pd

df = pd.DataFrame(student, columns=['Name','CH','MA','HI','CI','EN'], index = range(1,nums_int+1))
df
```

```python
df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
sum_series = df['CH'] + df['EN'] + df['MA'] + df['CI'] + df['HI']
sum_series
```

```python
df['總分'] = sum_series
df['平均'] = sum_series/5.0
df
```

```python
df['rank'] = df['平均'].rank(method='max')
df = df.sort_values(by=['rank'])
df
```
