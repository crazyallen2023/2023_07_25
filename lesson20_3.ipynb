```python
import yfinance as yf

data = yf.download('2330.TW', start="2022-01-01", end="2022-12-20")
tw2330 = data[['Adj Close']]
tw2330.columns = ["台積電"]
tw2330
```

```python
data = yf.download('2303.TW', start="2022-01-01", end="2022-12-20")
tw2303 = data[['Adj Close']]
tw2303.columns = ["聯電"]
tw2303
```

```python
data = yf.download('2454.TW', start="2022-01-01", end="2022-12-20")
tw2454 = data[['Adj Close']]
tw2454.columns = ["聯發科電"]
tw2454
```

```python
data = yf.download('2317.TW', start="2022-01-01", end="2022-12-20")
tw2317 = data[['Adj Close']]
tw2317.columns = ["鴻海"]
tw2317
```

```python
import pandas as pd

data2 = pd.concat([tw2330,tw2303,tw2454,tw2317],axis=1)
data2
```

```python
frame2 = data2.pct_change()
frame3 = frame2.fillna(0)
frame3
```

```python
frame3.corr()
```

```python
def fun1(v):
    return pd.Series([v.max(),v.min()],
                     index=["最高","最低"])

frame2.apply(fun1)
```

```python
# 換成百分比+字串差補
def fun1(v):
    max = v.max() * 100
    min = v.min() * 100
    return pd.Series([f'{max:.2f}%',f'{min:.2f}%'],
                     index=["最高","最低"])

frame2.apply(fun1)
```
