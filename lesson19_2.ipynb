```python
# serier's search
import pandas as pd
df = pd.read_csv('學生成績.csv')
```

```python
df.index = df['學號']
df
```

```python
df1 = df.drop(columns='學號')
df1
```

```python
df1['排名'] = df1['平均'].rank(method='min',ascending=False).astype(int)
df2 = df1.sort_values(by='排名')
df2
```

```python
#顯示前10跟後10名
display(df2.iloc[:10])
display(df2.iloc[-10:])

# df2.head(n=10)
# df2.tail(n=10)
```

```python
df2.query("排名<=10")
```

```python
#轉成EXCEL檔
df2.to_excel('學生成績單.xlsx',sheet_name='三年甲班')
```
