```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
font1=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

student1 = {'國文':80,'英文':70,'數學':99,'歷史':68,'探索':86}
student2 = {'國文':90,'英文':60,'數學':59,'歷史':98,'探索':66}
student3 = {'國文':67,'英文':57,'數學':61,'歷史':85,'探索':78}

figure = plt.figure(figsize=(8,5),dpi=200,facecolor='#cccccc')
axes1 = figure.add_subplot()

axes1.plot(student1.keys(), student1.values(),'r-.*',label='學生1')
student1_value_list = list(student1.values())
axes1.plot(student2.keys(), student2.values(),'b:+',label='學生2')
student2_value_list = list(student2.values())
axes1.plot(student3.keys(), student3.values(),'y:^',label='學生3')
student3_value_list = list(student3.values())

axes1.set_xticklabels(student1.keys(),fontproperties=font)

axes1.set_title("學生成績分布圖",fontproperties=font) #圖表名稱
axes1.set_xlabel("科目",fontproperties=font) #x軸名稱
axes1.set_ylabel("分數",fontproperties=font) #y軸名稱
axes1.grid(axis='y') #橫軸線
axes1.legend(prop=font) #說明欄

xloction = axes1.get_xticks()
xticks_count = len(xloction)

#座標點上的標記
for i in range(xticks_count):
    x = xloction[i]
    y1 = student1_value_list[i]
    y2 = student2_value_list[i]
    y3 = student3_value_list[i]

    axes1.annotate(str(y1),[x,y1])
    axes1.annotate(str(y2),[x,y2])
    axes1.annotate(str(y3),[x,y3])



plt.show()
```
