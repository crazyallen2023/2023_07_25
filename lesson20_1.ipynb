```python
# 直接用mean它是計算縱軸(x軸)的值
students_frame.mean()
```

```python
# 加上axis 可以計算每一個橫軸(x軸)的值
students_frame.mean(axis=1)
```

```python
# 這個方法沒有index 不推薦
for v in students_frame['CH']:
    if v >=90:
        print ('優')
    elif v>=80:
        print ('甲')
    elif v>=70:
        print ('乙')
    elif v>=60:
        print ('丙')
    else:
        print ('丁')
```

```python
def grade(v):
    if v>=90:
        return '優'
    elif v>=80:
        return '甲'
    elif v>=70:
        return '乙'
    elif v>=60:
        return '乙'
    else:
        return '丁'

students_frame['CH'].map(grade)
```

```python
# 把這個評比結果放回frame 用map(dunction)
students_frame['CH的評比'] = students_frame['CH'].map(grade)

students_frame
```

```python
students_frame = frame[['CH','MA','HI','CI','EN']]
students_frame
```

```python
# 取每科最低分者
def fun2(v):
    return v.min()

students_frame.apply(fun2)
```

```python
# 取每個人最低分項目
students_frame.apply(fun2, axis=1)
```

```python
# 取每個人不及格科數
def under60(v):
    return v[v<60].count()

students_frame.apply(under60, axis=1)
```

```python
students_frame['不及格科數'] = students_frame.apply(under60, axis=1)
students_frame
```
